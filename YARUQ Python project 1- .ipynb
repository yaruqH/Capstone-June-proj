```python
#(cleaning task) removing to NaN
missing_values = ["na","s"]
df =pd.read_csv("https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv",na_values=missing_values)
```

```python
df.head()
```

```python
#cleaning
# here i add a new column for Borough.  x is stand for bronx
df['Borough'] = df['School DBN'].str[2]
```

```python
# comparing the mean number scoring below 65 for all borough. MANH has a mean score below 65 of 1
# the measure i select to compare is the number scoring 65.
df2=df.groupby(['Borough'])['Number Scoring Below 65'].mean()
df2
```

```python

# creating Plot, with Mean score.
df2.plot(kind="bar",ylabel="Mean Number Scoring Below 65")
```
